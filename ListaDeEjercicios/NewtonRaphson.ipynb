```python
## Programa que calcula las raíces de una ecuación por el método de Newton-Raphson
## Elabora: Miguel Ángel Carrillo Lucía

## Primera función x**2 -2. Graficar

import numpy as np
import math
import matplotlib.pyplot as plt

x=np.linspace(-2,2,100)
y=(4*np.sin(x)-np.exp(x))
y1=(4*np.sin(x))
y2=(np.exp(x))
plt.figure(figsize=(8,6))
plt.title("Gráfica de la función ")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.plot(x, y1, label='f(x) = 4·sen(x)', color='blue')
plt.plot(x, y2, label='g(x) = e^x', color='red')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title("Gráfica de la función ")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.plot(x,y)
```

```python
def f(x):
    return 4*np.sin(x)-np.exp(x)

def df(x):
    return 4*np.cos(x)-np.exp(x)

# Parámetros iniciales
x0 = 0.32
tol = 1e-6
max_iter = 5

# Iteración de Newton-Raphson
for i in range(max_iter):
    x1 = x0 - f(x0)/df(x0)
    print(f"Iteración {i+1}: x = {x1:.6f}")
    if abs(x1 - x0) < tol:
        break
    x0 = x1

print(f"\nRaíz aproximada: {x1:.6f}")
```

```python
xr=x1
yr=f(xr)

plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x) = ', color='blue')
plt.axhline(0, color='black', linestyle='--')  # Eje x
plt.axvline(xr, color='gray', linestyle='--', label=f'x ≈ {xr:.4f}')
plt.plot(xr, yr, 'ro', label='Raíz aproximada')

plt.title("Visualización de la raíz de f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
