```python

from sympy import symbols, Eq, solve
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Ecuaciones
# x + y + z = 6
# 2x + y - z = 3
# -x + y + 2z = 14

x, y = np.linspace(-10, 10, 20), np.linspace(-10, 10, 20)
X, Y = np.meshgrid(x, y)

Z1 = 6 - X - Y
Z2 = (2*X + Y - 3)
Z3 = (14 + X - Y)/2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, alpha=0.5, color='red')
ax.plot_surface(X, Y, Z2, alpha=0.5, color='green')
ax.plot_surface(X, Y, Z3, alpha=0.5, color='blue')

# Punto de interseccin
from numpy.linalg import solve
A = np.array([[1, 1, 1], [2, 1, -1], [-1, 1, 2]])
b = np.array([6, 3, 14])
sol = solve(A, b)
ax.scatter(*sol, color='black', s=50, label="Solucin")
ax.legend()
plt.show()
```
